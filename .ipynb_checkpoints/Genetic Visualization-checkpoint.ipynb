```python
import numpy as np
import pandas as pd
```

```python
def transNUM(s):
    return list(map(float, eval(s)))
```

```python
resHash = {}
with open("Genetic.mi.log") as LOG:
    for line in LOG:
        fields = line.strip().split("\t")
        # Generate, MaxFitness, AvgFitness, MinFitness, Time
        resHash.setdefault(fields[4], {})
        GenNum = int(fields[0])
        maxFit = transNUM(fields[1])
        avgFit = transNUM(fields[2])
        for i in range(len(maxFit)):
            if i > 0:
                resHash[fields[4]].setdefault("Global-Avg", [])
                resHash[fields[4]].setdefault("Global-Max", [])
                resHash[fields[4]]["Global-Avg"].append(avgFit[i])
                resHash[fields[4]]["Global-Max"].append(maxFit[i])
            else:
                resHash[fields[4]].setdefault("Island%d-Avg" % i, [])
                resHash[fields[4]].setdefault("Island%d-Max" % i, [])
                resHash[fields[4]]["Island%d-Avg" % i].append(avgFit[i])
                resHash[fields[4]]["Island%d-Max" % i].append(maxFit[i])
print(resHash.keys())
print(resHash["1603437513"])
# df = pd.DataFrame(resHash["1603694358"])
# df
```
