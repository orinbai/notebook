```python
import numpy as np
import pandas as pd
from bokeh.io import output_notebook
import random
output_notebook()
from bokeh.plotting import figure, show
from bokeh.palettes import brewer
```

```python
def transNUM(s):
    return list(map(float, eval(s)))

def columnSLT(s, cname):
    return list(map(lambda x: x.startswith(s),cname))
```

```python
resHash = {}
with open("Genetic.mi.log") as LOG:
    for line in LOG:
        fields = line.strip().split("\t")
        # Generate, MaxFitness, AvgFitness, MinFitness, Time
        resHash.setdefault(fields[4], {})
        GenNum = int(fields[0])
        maxFit = transNUM(fields[1])
        avgFit = transNUM(fields[2])
        for i in range(len(maxFit)):
            if i == 0:
                resHash[fields[4]].setdefault("Avg-Global", [])
                resHash[fields[4]].setdefault("Max-Global", [])
                resHash[fields[4]]["Avg-Global"].append(avgFit[i])
                resHash[fields[4]]["Max-Global"].append(maxFit[i])
            else:
                resHash[fields[4]].setdefault("Avg-Island%d" % i, [])
                resHash[fields[4]].setdefault("Max-Island%d" % i, [])
                resHash[fields[4]]["Avg-Island%d" % i].append(avgFit[i])
                resHash[fields[4]]["Max-Island%d" % i].append(maxFit[i])
# print(resHash.keys())
# print(resHash["1603437513"])
# df = pd.DataFrame(resHash["1603712383"])
df = pd.DataFrame(resHash["1603781856"])
```
